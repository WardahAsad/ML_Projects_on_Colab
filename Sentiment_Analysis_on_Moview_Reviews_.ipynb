```python
import pandas as pd
import nltk
nltk.download('punkt')
```

```python
def tokenize_text(text):
  return[word for sentence in nltk.sent_tokenize(text) for word in nltk.word_tokenize(sentence)]

text = """ Pakistan, officially the Islamic Republic of Pakistan, is a diverse country located in South
Asia. It is the fifth-most populous country globally, with a rich history and cultural
heritage. Known for its breathtaking landscapes, from the mountains of the north to the
coastal areas in the south, Pakistan is a land of contrasts. The country faces various
challenges, including economic instability and social issues, but its resilient people
continue to strive for progress. With a growing economy and a young population,
Pakistan holds potential for a brighter future."""

print(tokenize_text(text))
print(len(tokenize_text(text)))
```

```python
nltk.download('stopwords')

from nltk.corpus import stopwords
sw_english = stopwords.words('english')
```

```python
def remove_stopwords(text):
  words = [word for word in tokenize_text(text) if word.lower() not in sw_english]
  return " ".join(words)

removed_stopwords = remove_stopwords(text)
print(removed_stopwords)
print("Old Length = ", len(text))
print("New Length = ", len(remove_stopwords(text)))
```

```python
def make_lowercase(text):
  words = [word.lower() for word in tokenize_text(text)]
  return " ".join(words)

lowercased = make_lowercase(removed_stopwords)
print(lowercased)
```

```python
stemmer = nltk.stem.PorterStemmer()

def stemming(text):
  stems = [stemmer.stem(word) for word in tokenize_text(text)]
  return " ".join(stems)

stemmed_text = stemming(lowercased)
print(stemmed_text)
```

```python
nltk.download('wordnet')

lemmatizer = nltk.stem.WordNetLemmatizer()

def lemmatization(text):
  lemma = [lemmatizer.lemmatize(word) for word in tokenize_text(text)]
  return " ".join(lemma)

print(lemmatization(lowercased))
```

```python
nltk.download('movie_reviews')
```

```python
from nltk.corpus import movie_reviews
```

```python
review_text = []
review_target = []

for category in movie_reviews.categories():
  for fileid in movie_reviews.fileids(category):
    review_text.append(movie_reviews.raw(fileid))
    review_target.append(category)

len(review_text)
```

```python
print(review_text)
```

```python
movie_reviews_df = pd.DataFrame(review_text)
movie_reviews_df['target'] = review_target
movie_reviews_df.columns = ['review', 'sentiment']
movie_reviews_df['sentiment'].replace({'pos' : 1, 'neg' : 0 }, inplace = True)
movie_reviews_df
```

```python
# using Bag of words
from sklearn.model_selection import train_test_split

X = movie_reviews_df['review']
y = movie_reviews_df['sentiment']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)

vectorizer = CountVectorizer(lowercase=True, stop_words='english')

X_train = vectorizer.fit_transform(X_train)

X_test = vectorizer.transform(X_test)

print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

from sklearn.model_selection import train_test_split

X = movie_reviews_df['review']
y = movie_reviews_df['sentiment']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)

vectorizer = TfidfVectorizer(lowercase = True , stop_words = 'english')
X_train = vectorizer.fit_transform(X_train)
X_test =vectorizer.transform(X_test)

print(X_train.shape)
print(X_test.shape)
```

```python
# training model using Naive Bayes

from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()

model.fit(X_train, y_train)


accuracy = model.score(X_test, y_test)
print("Accuracy using Naive Bayes = ", accuracy)
```

```python
#prediction using bow
def predict_sentiment(review):
  review_bow=vectorizer.transform([review])
  pred=model.predict(review_bow)
  return 'Positive' if pred[0]==1 else 'Negative'

user_review=input("Enter your review:")
print("Sentiment Predicted by Model is:",predict_sentiment(user_review))
```

```python
# SVM
from sklearn.svm import SVC
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score

from sklearn.model_selection import train_test_split

X = movie_reviews_df['review']
y = movie_reviews_df['sentiment']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)

vectorizer = TfidfVectorizer(lowercase = True , stop_words = 'english')
X_train = vectorizer.fit_transform(X_train)
X_test =vectorizer.transform(X_test)

# Initialize SVM classifier
svm_model = SVC(kernel='linear')  # You can change the kernel as needed (e.g., 'linear', 'rbf')

# Train the SVM model
svm_model.fit(X_train, y_train)

# Calculate accuracy
accuracy = accuracy_score(y_test, svm_model.predict(X_test))
print("Accuracy using SVM =", accuracy)

# Prediction function using TF-IDF
def predict_sentiment(review):
    review_tfidf = vectorizer.transform([review])
    pred = svm_model.predict(review_tfidf)
    return 'Positive' if pred[0] == 1 else 'Negative'

# Example usage:
user_review = input("Enter your review:")
print("Sentiment Predicted by Model is:", predict_sentiment(user_review))
```

```python
# Decision Tree
from sklearn.tree import DecisionTreeClassifier

dtr = DecisionTreeClassifier()
dtr.fit(X_train, y_train)
accuracy = dtr.score(X_test, y_test)
print("The accuracy of Decision Tree is", accuracy, "\n")

#prediction using TfIDF
def predict_sentiment(review):
  review_tfidf=vectorizer.transform([review])
  pred=dtr.predict(review_tfidf)
  return 'Positive' if pred[0]==1 else 'Negative'
user_review=input("Enter your review:")
print("Sentiment Predicted by Model is:",predict_sentiment(user_review))
```
